```python
#Dataset uploading
import pandas as pd

# /content/raw_data.csv
file_path = '/content/train.csv'

# Data Loading using Pandas
df = pd.read_csv(file_path)
```

```python
#Data set preview
df.head(10)
```

```python
df.info()
df.describe()
```

```python
clean_data = df.copy()
cd = pd.DataFrame(df)
cd.head()
```

#In the given data we can see many unit values like kmpl , CC , bhp for columns like mileage , Power except , Engine ,  we will now extract the decimals for the columns.

```python
cd['Mileage'] =  cd['Mileage'].str.extract('(\d+\.\d+)').astype(float)
cd['Engine'] =  cd['Engine'].str.replace(' CC', '').astype(float)
cd['Power'] =  cd['Power'].str.extract('(\d+\.\d+)').astype(float)
cd['New_Price'] =  cd['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
cd.head(10)
```

#now we can see data without units the data consists of numerical or alphabets

```python
# checking for missing values.
missing_values = cd.isnull().sum()
print("Missing Values count :")
print(missing_values)
```

#there are certain fields which contains null values , those can be replaced them with the mean or medians

```python
#for numerical values , replacing missing values with mean
cd['Mileage'].fillna(cd['Mileage'].mean(), inplace=True)
cd['Engine'].fillna(cd['Engine'].mean(), inplace=True)
cd['Power'].fillna(cd['Power'].mean(), inplace=True)
cd['Seats'].fillna(cd['Seats'].mean(), inplace=True)
cd['New_Price'].fillna(cd['New_Price'].mean(), inplace=True)

#for categorical values , replacing missing values with mean
cd['Location'].fillna(cd['Location'].mode()[0], inplace=True)
cd['Fuel_Type'].fillna(cd['Fuel_Type'].mode()[0], inplace=True)
cd['Transmission'].fillna(cd['Transmission'].mode()[0], inplace=True)
```

```python
cd['Power'] = cd['Power'].round(2)
cd['New_Price'] = cd['New_Price'].round(2)
cd['Mileage'] = cd['Mileage'].round(2)


cd.to_csv('cleaned_dataset.csv', index=False)
```

```python
missing_values = cd.isnull().sum()
print("Missing Values count After ")
print(missing_values)
```

```python
#Convert Categorical Variables into One-Hot Encoded Values
cd1 = pd.get_dummies(cd, columns=['Fuel_Type', 'Transmission'], drop_first=True)

cd1.head(10)

cd1.to_csv('cd1.csv',index=False)
```

```python
# Create a New Feature (Current Age of the Car):

current_year = 2024  # Current year
cd1['Current_Age'] = current_year - cd1['Year']
```

```python
# Selecting the specific columns
selected_df = cd[['Name', 'Fuel_Type', 'Engine']]
print(selected_df)
```

```python
# Filtering the data based on conditions
filtered = cd1[cd['Fuel_Type'] == 'Petrol']
print(filtered)
```

```python
renamed = cd1.rename(columns={'Transmission': 'Transmission_Type', 'Seats': 'Seats_No'})
print(renamed)
```

```python
# Mutating data (creating a new calculated column)
cd1['Price_Category'] = pd.cut(cd1['Price'], bins=[0, 5, 10, 15, 20, float('inf')], labels=['0-5', '5-10', '10-15', '15-20', '20+'])
print(cd1)

cd1.head(5)
```

```python
# Arrange the data
arranged = cd1.sort_values(by='Price', ascending=False)
print(arranged)
```

```python
# Summarizing with group by operations
summary = cd1.groupby('Name').agg({'Price': 'mean', 'Mileage': 'median'})

print(summary)
```
